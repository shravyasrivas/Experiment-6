```python
#7. Write a program to print the next date
dd = int(input("enter date:"))
mm = int(input("enter month:"))
yyyy = int(input("enter year:"))

# Validate year
if yyyy <= 0:
    print("Invalid year. Please enter a positive year.")
elif not (1 <= mm <= 12):
    print("Invalid month. Please enter a month between 1 and 12.")
else:
    # Determine the number of days in the given month
    if mm == 2: # February
        if (yyyy % 4 == 0 and yyyy % 100 != 0) or (yyyy % 400 == 0):
            max_days = 29
        else:
            max_days = 28
    elif mm in [4, 6, 9, 11]: # April, June, September, November
        max_days = 30
    else: # January, March, May, July, August, October, December
        max_days = 31

    # Validate day
    if not (1 <= dd <= max_days):
        print(f"Invalid day for month {mm} in year {yyyy}. Please enter a day between 1 and {max_days}.")
    else:
        # Calculate the next date
        dd += 1
        if dd > max_days:
            dd = 1
            mm += 1
            if mm > 12:
                mm = 1
                yyyy += 1
        print(f"The next date is: {dd}/{mm}/{yyyy}")
```
